## Task-4 Terrorism (Data Analysis)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
df=pd.read_csv("datasets/terrorism.csv",encoding="latin",low_memory=False)
df.head(5)
```

```python
df.info()
```

```python
data=df.copy()
```

## Handling NULL Values

```python
data.isna().sum()
```

```python
nan_features=[feature for feature in df.columns  if df[feature].isna().sum()>100000]
```

```python
na_data=df[nan_features]
```

```python
data.drop(nan_features,axis=1,inplace=True)
data.head()
```

### Deleting Columns that Don't Make sense

```python
data.drop(["eventid","INT_LOG","INT_IDEO","INT_MISC",'INT_ANY',"claimed","summary","dbsource","scite1"],axis=1,inplace=True)
```

```python
data.drop(["crit1","crit2","crit3","latitude","longitude"],axis=1,inplace=True)
data.head()
```

```python
for feature in data.columns:
    if feature+"_txt" in data.columns :
        data.drop(feature,axis=1,inplace=True)
```

```python
data.isna().sum()
```

### Some features have Negative Values: Fixing them 

```python
for feature in data.columns:
    if data[feature].dtype!='O':
        data[feature]=np.abs(data[feature])
```

## Imputing NA values 

```python
NA=data.dropna(axis=0)
```

```python
data.multiple.fillna(0.0,inplace=True)

data.doubtterr.fillna(0.0,inplace=True)

data.guncertain1.fillna(0.0,inplace=True)

data.ishostkid.fillna(0,inplace=True)

data.nwoundte.fillna(np.median(NA.nwoundte),inplace=True)
data.nwoundus.fillna(np.median(NA.nwoundus),inplace=True)
data.nwound.fillna(np.median(NA.nwound),inplace=True)

data.nkill.fillna(np.median(NA.nkill),inplace=True)
data.nkillus.fillna(np.median(NA.nkillus),inplace=True)
data.nkillter.fillna(np.median(NA.nkillter),inplace=True)

data.nperpcap.fillna(np.median(NA.nperpcap),inplace=True)
data.nperps.fillna(np.median(NA.nperps),inplace=True)

data['specificity'].fillna(np.median(NA["specificity"]),inplace=True)
```

```python
data.drop(["targsubtype1_txt"],axis=1,inplace=True)
data.isna().sum()
```

```python
data.drop("weapdetail",inplace=True,axis=1)
```

```python
data.doubtterr.replace(9.0,0.0,inplace=True)
data.ishostkid.replace(9.0,0.0,inplace=True)
```

### Converting Data Type of some Feature

```python
for feature in data.columns:
    if data[feature].dtype=='int32':
        data[feature]=data[feature].astype("int64")
        
```
